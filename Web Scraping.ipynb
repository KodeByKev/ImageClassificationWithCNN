```python
import requests

def google_image_search(query, se_id, se_key, num_results=200):
    search_params = {
        'cx': se_id,
        'key': se_key,
        'filter': 1,
        'imgType': 'photo',
        'num': 10,
        'searchType': 'image',
        'q': query
    }

    image_urls = []
    try:
        start = 1
        while len(image_urls) < num_results:
            search_params['start'] = start
            r = requests.get('https://customsearch.googleapis.com/customsearch/v1', params=search_params)
            r.raise_for_status()
            items = r.json().get('items', [])
            if not items:
                break
            
            image_urls.extend([item['link'] for item in items if 'link' in item])
            start += len(items)

        return image_urls[:num_results]  

    except requests.exceptions.RequestException as e:
        print("Error:", e)
        return []

se_id = '66d42d2512af44d83'
se_key = 'AIzaSyDgfLDQr5J2MOnJrwhuf-IIA5w-HFy77YM'

query = input("Search Query:")

image_urls = google_image_search(query, se_id, se_key)

if image_urls:
    print("Results:")
    for url in image_urls:
        print(url)
else:
    print("No Results")
```
